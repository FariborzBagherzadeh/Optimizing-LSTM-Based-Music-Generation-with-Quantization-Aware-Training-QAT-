```python
import os
import tensorflow as tf
from tensorflow.python.keras.layers import Input, Dense, Lambda
from tensorflow.python.keras.layers.recurrent import LSTM
from tensorflow.python.keras.models import Model
import numpy as np
import pretty_midi
import librosa
import soundfile as sf
np.int = int
```

```python
# Path of the Midi Files
folder_path = os.path.join(os.getcwd(),'midi_files')

def load_midi_files(folder_path):
    midi_data =[]
    for file in os.listdir(folder_path):
        if file.endswith('.mid'):
            midi = pretty_midi.PrettyMIDI(os.path.join(folder_path, file))
            piano_roll = midi.get_piano_roll(fs=100).T  # Time resolution = 100 Hz
            midi_data.append(piano_roll)
    print(midi_data)

def preprocess_midi_data(midi_data, max_len=1000):
    processed_data = []
    for roll in midi_data:
        roll = roll[:max_len, :]
        roll = roll / np.max(roll)
load_midi_files(folder_path)
```
